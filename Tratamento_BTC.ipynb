### Tratamento da base de dados de BTC

```python
import pandas as pd
```

```python
btc = pd.read_csv('bitstampUSD_1-min_data_2012-01-01_to_2021-03-31.csv')
```

```python
btc.head()
```

```python
# Convertendo de Unix para datetime:
btc['Date'] = pd.to_datetime(btc['Timestamp'], unit = 's')
```

```python
btc.head()
```

```python
btc.sort_values('Date',inplace=True)
btc.tail()
```

```python
# Removendo colunas desnecessárias e redefinindo index:
btc = btc [['Date','Open','High','Low','Close','Volume_(BTC)','Volume_(Currency)']]
btc.set_index('Date', drop=True,inplace=True)
```

```python
# Dicionário para definir critérios de agregação ao agrupar por dia:
ohlc_dict = 
{                                                                                                             
    'Open': 'first',                                                                                                    
    'High': 'max',                                                                                                       
    'Low': 'min',                                                                                                        
    'Close': 'last',                                                                                                    
    'Volume_(BTC)': 'sum',
    'Volume_(Currency)': 'sum',
}

btc_daily = btc.resample('D').agg(ohlc_dict)
btc_daily.head()
```

```python
# Temos dados de 2011 - 2021
# Precisamos apenas de 2015 a 2019 (período dos comentários)
btc_daily = btc_daily.loc['2015-01-01':'2019-12-31']
```

```python
btc_daily.to_csv('btc_daily.csv')
```
